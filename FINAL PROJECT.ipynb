```python
dc_listings = pd.read_csv('C:\\Users\hp\Desktop\dc_airbnb.csv')
dc_listings = dc_listings[['accommodates','bedrooms','bathrooms','beds','price','minimum_nights','maximum_nights','number_of_reviews']]
dc_listings['price'] = dc_listings.price.str.replace("\$|,",'').astype(float)
dc_listings = dc_listings[np.isfinite(dc_listings['bedrooms'])]
dc_listings = dc_listings[np.isfinite(dc_listings['beds'])]
dc_listings = dc_listings[np.isfinite(dc_listings['bathrooms'])]

print (dc_listings.shape)
dc_listings.head()
```

### KNN ALGORITHM

```python
from sklearn.neighbors import KNeighborsRegressor
# Using the Second Machine Learning Algorithm to create the model : K-Nearest Neighbors Regressor 
knn = KNeighborsRegressor(algorithm='brute')
normalized_listings = dc_listings.sample(frac=1,random_state=0)
norm_train_df = normalized_listings.copy().iloc[0:2792]
norm_test_df = normalized_listings.copy().iloc[2792:]

cols = ['accommodates','bedrooms','bathrooms','beds']
# Training our data to fit the Model
knn.fit(norm_train_df[cols], norm_train_df['price'])
#Predicting the outcome from the test data
four_features_predictions = knn.predict(norm_test_df[cols])
from sklearn.metrics import mean_squared_error
# Calculating the error we have from the Model prediction from the actual value of the test data
four_features_mse = mean_squared_error(norm_test_df['price'], four_features_predictions)
four_features_rmse = four_features_mse ** (1/2)
four_features_rmse
```

### RANDOM FOREST

```python
from sklearn.ensemble import RandomForestRegressor
# Using the Third Machine Learning Algorithm to create the model : Random Forest Regressor
# Instantiate model with 1000 decision trees
# These parameters can be twicked to get the best Model for this DataSet
rf = RandomForestRegressor(n_estimators = 100, random_state = 42)
cols = ['accommodates','bedrooms','bathrooms','beds']
# Training our data to fit the Model
rf.fit(norm_train_df[cols], norm_train_df['price'])
#Predicting the outcome from the test data
four_features_predictions = rf.predict(norm_test_df[cols])
from sklearn.metrics import mean_squared_error
# Calculating the error we have from the Model prediction from the actual value of the test data
four_features_mse = mean_squared_error(norm_test_df['price'], four_features_predictions)
four_features_rmse = four_features_mse ** (1/2)
four_features_rmse
```
